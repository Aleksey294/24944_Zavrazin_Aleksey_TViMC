```python
# lib
import matplotlib.pyplot as plt
import random
import math
```

```python
# task_1
def frequency_probability(trials=10000):
    """
    Бросаем монету trials раз и считаем вероятность выпадения орла
    """
    heads = 0
    for _ in range(trials):
        if random.random() < 0.5:  # равновероятная монета
            heads += 1
    probability = heads / trials
    print(f"Частотная вероятность выпадения орла при {trials} испытаниях: {probability:.4f}")

frequency_probability()
```

```python
def monte_carlo_pi(trials=10000):
    inside_circle = 0
    pi_estimates = []

    for i in range(1, trials + 1):
        x, y = random.random(), random.random()
        if x**2 + y**2 <= 1:
            inside_circle += 1
        pi_estimate = 4 * inside_circle / i
        pi_estimates.append(pi_estimate)

    # финальный результат
    print(f"Приближённое значение π после {trials} испытаний: {pi_estimates[-1]:.6f}")

    # визуализация
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, trials + 1), pi_estimates, label="Приближение π (Монте-Карло)", color="blue")
    plt.axhline(y=math.pi, color="red", linestyle="--", label=f"Истинное π = {math.pi:.6f}")
    plt.xlabel("Число испытаний")
    plt.ylabel("Значение π")
    plt.title("Сходимость метода Монте-Карло к числу π")
    plt.legend()
    plt.grid(True)
    plt.show()

monte_carlo_pi()
```

```python
# task_3
class LCG:
    def __init__(self, seed=1, a=6364136223846793005, c=1, m=2**64):
        self.state = seed
        self.a = a
        self.c = c
        self.m = m

    def random(self):
        self.state = (self.a * self.state + self.c) % self.m
        return self.state / self.m



def monte_carlo_pi_lcg(trials=1000000):

    rng = LCG(seed=42)
    inside_circle = 0

    for _ in range(trials):
        x, y = rng.random(), rng.random()
        if x ** 2 + y ** 2 <= 1:
            inside_circle += 1

    pi_estimate = 4 * inside_circle / trials
    print(f"[LCG] Приближённое значение π после {trials} испытаний: {pi_estimate:.6f}")

monte_carlo_pi_lcg()
```
