```python
# lib
import matplotlib.pyplot as plt
import random
import math
```

```python
# task_1
def frequency_probability(trials=10000):
    """
    Бросаем монету trials раз и считаем вероятность выпадения орла
    """
    heads = 0
    for _ in range(trials):
        if random.random() < 0.5:  # равновероятная монета
            heads += 1
    probability = heads / trials
    print(f"Частотная вероятность выпадения орла при {trials} испытаниях: {probability:.4f}")

frequency_probability()
```

```python
# task_2
def monte_carlo_pi(trials=100000):
    inside_circle = 0
    x_vals, y_vals = [], []

    for _ in range(trials):
        x, y = random.random(), random.random() 
        if x ** 2 + y ** 2 <= 1:
            inside_circle += 1
            x_vals.append(x)
            y_vals.append(y)

    pi_estimate = 4 * inside_circle / trials
    print(f"Приближённое значение π после {trials} испытаний: {pi_estimate:.6f}")

    # визуализация
    plt.figure(figsize=(5, 5))
    circle = plt.Circle((0, 0), 1, color="r", fill=False)
    plt.gca().add_patch(circle)
    plt.scatter(x_vals, y_vals, s=1, color="blue")
    plt.title(f"Приближение числа π: {pi_estimate:.6f}")
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

monte_carlo_pi()
```

```python
# task_3
class LCG:
    def __init__(self, seed=1, a=1664525, c=1013904223, m=2**32):
        self.state = seed
        self.a = a
        self.c = c
        self.m = m

    def random(self):
        self.state = (self.a * self.state + self.c) % self.m
        return self.state / self.m


def monte_carlo_pi_lcg(trials=100000):

    rng = LCG(seed=42)
    inside_circle = 0

    for _ in range(trials):
        x, y = rng.random(), rng.random()
        if x ** 2 + y ** 2 <= 1:
            inside_circle += 1

    pi_estimate = 4 * inside_circle / trials
    print(f"[LCG] Приближённое значение π после {trials} испытаний: {pi_estimate:.6f}")

monte_carlo_pi_lcg()
```
